```python
from matplotlib import pyplot as plt

# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='red', linestyle='--',
             marker='o', markerfacecolor='yellow',
             ecolor='black',capsize=5,
             label='Y values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right')
plt.show()
```

```python
import numpy as np
x = np.linspace(0,50,200)
y=np.exp(np.negative(x)/10)*np.sin(x)

title='y = $\\ e^{-\\frac{x}{10}}\\sin(x)$'

plt.plot(x,y)
plt.xlabel('X',color='darkorange',fontsize=15)
plt.ylabel('Y',color='darkorange',rotation=0,fontsize=15)
plt.title(title,fontsize=20)
plt.grid(alpha=0.25)
plt.ylim(-1,1)
```

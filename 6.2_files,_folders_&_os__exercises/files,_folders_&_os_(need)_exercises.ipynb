```python
import os
import glob
import shutil
#task 1 and 4 done together (oops)
mth_to_no = {
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}

def renameddmm(files):
    newfiles=[]
    for i,file in enumerate(files):
        oldname = os.path.split(files[i])[-1]
        name = oldname.split('_')
        date = name[0].split('-')
        if date[0] in mth_to_no:
            date[0] = mth_to_no[date[0]]
        try:
            newname = f'{date[0]}-{date[1]}-{name[1]}'
            newfiles.extend([newname])
        except:
            newfiles.extend([oldname]) #assuming files are renamed correctly already
            continue
    return newfiles

#1 access files, #2 check names, #3 change names as required
#C:\Users\65889\Documents\GitHub\learning-portfolio-2024-aug-Arcen3\6.2_files,_folders_&_os__exercises\os-collaboration-exercise-data

path = os.path.join('.','os-collaboration-exercise-data')
files = glob.glob(os.path.join('os-collaboration-exercise-data','*'))
correctednames = renameddmm(files)
for i,files in enumerate(files):
    print(files)
    oldpath = os.path.join(files)
    newpath = os.path.join(path,correctednames[i])
    if oldpath!=newpath:
        os.rename(oldpath,newpath)
    else:
        continue
    
```
